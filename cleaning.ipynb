added breaks and created a separate txt for chapters

```python
import os

# Path to your cleaned text file
input_file_path = "C:\\Users\\ananya\\Desktop\\doctor\\ouput\\cleaned_document2.txt"
output_directory = "C:\\Users\\ananya\\Desktop\\doctor\\ouput\\chapters"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Separator to use
separator = "=" * 50 + "\n" + "■ FURTHER READING" + "\n" + "=" * 50 + "\n\n"

def save_chapter(chapter_number, content):
    filename = f"chapter_{chapter_number:03d}.txt"
    filepath = os.path.join(output_directory, filename)
    with open(filepath, 'w', encoding='utf-8') as file:
        file.write(content)
    print(f"Saved {filename}")

# Read the cleaned document and process it
with open(input_file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Split the content into chapters
chapters = content.split("■ FURTHER READING")

# Process and save each chapter
for i, chapter in enumerate(chapters, 1):
    if i == 1:
        # First chapter doesn't need a separator at the beginning
        formatted_chapter = chapter.strip() + "\n\n" + separator
    else:
        formatted_chapter = separator + chapter.strip() + "\n\n" + separator
    
    save_chapter(i, formatted_chapter)

print(f"Split {len(chapters)} chapters into separate files in {output_directory}")
```

cleaned the txt files

```python
import os
import re

# Directory containing the chapter files
input_directory = "C:\\Users\\ananya\\Desktop\\doctor\\ouput\\chapters"
output_directory = "C:\\Users\\ananya\\Desktop\\doctor\\ouput\\cleaned_chapters1"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Patterns to match
separator_pattern = r"={50,}\s*■ FURTHER READING\s*={50,}"
end_pattern = r"---"
further_reading_pattern = r"\n?={50,}\n■ FURTHER READING\n={50,}\n?"  # Match entire further reading section with newlines

def clean_chapter(content):
    # Find the first occurrence of the separator pattern
    separator_match = re.search(separator_pattern, content)
    if separator_match:
        start_index = separator_match.start()
        # Find the first occurrence of "---" after the separator
        remaining_content = content[start_index:]
        end_match = re.search(end_pattern, remaining_content)
        if end_match:
            end_index = start_index + end_match.end()
            # Remove the content between the separator and the first "---"
            content = content[:start_index] + content[end_index:]

    # Remove the "■ FURTHER READING" section with its surrounding separators, if it exists
    content = re.sub(further_reading_pattern, '', content)

    return content

# Process each file in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".txt"):
        input_path = os.path.join(input_directory, filename)
        output_path = os.path.join(output_directory, filename)
        
        with open(input_path, 'r', encoding='utf-8') as file:
            content = file.read()
        
        cleaned_content = clean_chapter(content)
        
        with open(output_path, 'w', encoding='utf-8') as file:
            file.write(cleaned_content)
        
        print(f"Cleaned and saved: {filename}")

print(f"All files processed. Cleaned files are in {output_directory}")
```

mapping to yaml file and categorizing into folders

```python
import os
import re
import shutil
import yaml

# Paths
input_directory = "C:\\Users\\ananya\\Desktop\\doctor\\ouput\\cleaned_chapters1"
output_directory = "C:\\Users\\ananya\\Desktop\\doctor\\ouput\\final_structure"

# Load the folder structure from the YAML file
yaml_file_path = "C:\\Users\\ananya\\Desktop\\doctor\\folder_structure.yaml"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Load YAML configuration for folder structure
with open(yaml_file_path, 'r', encoding='utf-8') as file:
    folder_structure = yaml.safe_load(file)

# Function to sanitize folder names for Windows (replace special characters)
def sanitize_folder_name(folder_name):
    # Replace colon, parentheses, and other invalid characters
    return re.sub(r'[<>:"/\\|?*]', '', folder_name)

# Function to find the correct folder based on the chapter number
def get_folder_for_chapter(chapter_number):
    for main_folder, subfolders in folder_structure.items():
        sanitized_main_folder = sanitize_folder_name(main_folder)
        for subfolder, chapter_range in subfolders.items():
            sanitized_subfolder = sanitize_folder_name(subfolder)
            if chapter_number in chapter_range:
                return os.path.join(output_directory, sanitized_main_folder, sanitized_subfolder)
    return None

# Function to extract chapter number from filename
def extract_chapter_number(filename):
    # Assuming chapter filenames are like "chapter_13.txt", "chapter_144.txt", etc.
    match = re.search(r"(\d+)", filename)
    if match:
        return int(match.group(1))
    return None

# Process each file in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".txt"):
        input_path = os.path.join(input_directory, filename)
        
        # Extract chapter number from the filename
        chapter_number = extract_chapter_number(filename)
        
        if chapter_number is not None:
            # Get the target folder based on the chapter number
            target_folder = get_folder_for_chapter(chapter_number)
            
            if target_folder:
                # Create the target folder if it doesn't exist
                os.makedirs(target_folder, exist_ok=True)
                
                # Move the file to the appropriate folder
                target_path = os.path.join(target_folder, filename)
                shutil.move(input_path, target_path)
                
                print(f"Moved {filename} to {target_folder}")
            else:
                print(f"Chapter {chapter_number} not found in folder structure.")
        else:
            print(f"Could not extract chapter number from {filename}")

print(f"All files have been processed and moved to {output_directory}")
```
